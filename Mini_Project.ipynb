```python
def contentTokenizer(content) :
    contenttoken = []
    sym = "~!@#$%^&*()_+-=`{}[]|\\:;\"',./<>?"
    for x in range(len(content[0])) :
        new = content[0][x].lower()
        for s in sym :
            new = new.replace(s, " ")
        contenttoken.extend(new.split())
    return contenttoken
```

```python
def makeline2(txt) :
    txt2 = []
    for x in range(len(txt)) :
        txt1 = txt[x].replace("\"", '').split("\n")
        while '' in txt1 :
            txt1.remove('')
        txt2.extend(txt1)
    return txt2
```
